## Import required libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import os
pd.set_option('display.max_rows', 500)
import requests
import json
from pandas.io.json import json_normalize
```

## API's
#### It was nos possible to get all 2019 registered crimes in just one request, so we'll request month by month and then concatenate the dataframes.

```python
# headers definition
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.142 Safari/537.36"}

#2019 January
url1 = 'https://data.police.uk/api/crimes-street/all-crime?poly=53.496965,%20-2.276478:53.480613,%20-2.194617:53.462255,%20-2.265110&date=2019-01'
response1 = requests.get(url1, headers = headers)
results1 = response1.json()
flattened1 = json_normalize(results1)

#2019 February
url2 = 'https://data.police.uk/api/crimes-street/all-crime?poly=53.496965,%20-2.276478:53.480613,%20-2.194617:53.462255,%20-2.265110&date=2019-02'
response2 = requests.get(url2, headers = headers)
results2 = response2.json()
flattened2 = json_normalize(results2)

#2019 March
url3 = 'https://data.police.uk/api/crimes-street/all-crime?poly=53.496965,%20-2.276478:53.480613,%20-2.194617:53.462255,%20-2.265110&date=2019-03'
response3 = requests.get(url3, headers = headers)
results3 = response3.json()
flattened3 = json_normalize(results3)

#2019 April
url4 = 'https://data.police.uk/api/crimes-street/all-crime?poly=53.496965,%20-2.276478:53.480613,%20-2.194617:53.462255,%20-2.265110&date=2019-04'
response4 = requests.get(url4, headers = headers)
results4 = response4.json()
flattened4 = json_normalize(results4)

#2019 May
url5 = 'https://data.police.uk/api/crimes-street/all-crime?poly=53.496965,%20-2.276478:53.480613,%20-2.194617:53.462255,%20-2.265110&date=2019-05'
response5 = requests.get(url5, headers = headers)
results5 = response5.json()
flattened5 = json_normalize(results5)

#2019 June
url6 = 'https://data.police.uk/api/crimes-street/all-crime?poly=53.496965,%20-2.276478:53.480613,%20-2.194617:53.462255,%20-2.265110&date=2019-06'
response6 = requests.get(url6, headers = headers)
results6 = response6.json()
flattened6 = json_normalize(results6)
```

```python
#2018 January
url18_1 = 'https://data.police.uk/api/crimes-street/all-crime?poly=53.496965,%20-2.276478:53.480613,%20-2.194617:53.462255,%20-2.265110&date=2018-01'
response181 = requests.get(url18_1, headers = headers)
results181 = response181.json()
flattened181 = json_normalize(results181)

#2018 February
url18_2 = 'https://data.police.uk/api/crimes-street/all-crime?poly=53.496965,%20-2.276478:53.480613,%20-2.194617:53.462255,%20-2.265110&date=2018-02'
response182 = requests.get(url18_2, headers = headers)
results182 = response182.json()
flattened182 = json_normalize(results182)

#2018 March
url18_3 = 'https://data.police.uk/api/crimes-street/all-crime?poly=53.496965,%20-2.276478:53.480613,%20-2.194617:53.462255,%20-2.265110&date=2018-03'
response183 = requests.get(url18_3, headers = headers)
results183 = response183.json()
flattened183 = json_normalize(results183)

#2018 April
url18_4 = 'https://data.police.uk/api/crimes-street/all-crime?poly=53.496965,%20-2.276478:53.480613,%20-2.194617:53.462255,%20-2.265110&date=2018-04'
response184 = requests.get(url18_4, headers = headers)
results184 = response184.json()
flattened184 = json_normalize(results184)

#2018 May
url18_5 = 'https://data.police.uk/api/crimes-street/all-crime?poly=53.496965,%20-2.276478:53.480613,%20-2.194617:53.462255,%20-2.265110&date=2018-05'
response185 = requests.get(url18_5, headers = headers)
results185 = response185.json()
flattened185 = json_normalize(results185)

#2018 June
url18_6 = 'https://data.police.uk/api/crimes-street/all-crime?poly=53.496965,%20-2.276478:53.480613,%20-2.194617:53.462255,%20-2.265110&date=2018-06'
response186 = requests.get(url18_6, headers = headers)
results186 = response186.json()
flattened186 = json_normalize(results186)
```

```python
frames = [flattened1, flattened2, flattened3, flattened4, flattened5, flattened6]
frames2 = [flattened181, flattened182, flattened183, flattened184, flattened185, flattened186]
data = pd.concat(frames)
data18 = pd.concat(frames2)
```
