```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import urllib.request
from urllib.request import urlopen
import re
```

```python
url = 'https://www.playgroundmag.net/'

html = requests.get(url).content

soup = BeautifulSoup(html, 'lxml')

soup
```

```python
lst = soup.find_all('p', {'class': 'title'}) 

lst
```

```python
titles = [title.get_text() for title in lst]

titles
```

```python
playground_string = "\n".join(titles)

print(playground_string)
```

```python
save_file = open('output/playground.txt', 'w')
save_file.write(playground_string)
save_file.close()
```
